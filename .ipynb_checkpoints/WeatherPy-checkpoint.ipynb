```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import urllib
from scipy.stats import linregress

#Import API key
from weather_api_key import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
city = "yellowknife"
query_url = url + "appid=" + weather_api_key + "&q=" + city
json_output = requests.get(query_url).json()
json_output
```

```python
lat=[]
long=[]
temp=[]
cloudiness=[]
winds = []
country=[]
date=[]
humid=[]
city_name=[]
max_temp=[]
url = "http://api.openweathermap.org/data/2.5/weather?"
units= "metrics"
#query_url = url + units + weather_api_key
query_url = f"{url}appid={weather_api_key}&units={units}&q="
#counter 2 is for sets of 50
#counter 1 is the counting of 1-50
counter1 = 1
counter2 =1

for i, city in enumerate(cities[:600]):
    if (i %50 ==0 and i>=50):
        counter2 += 1
        counter1 = 0
    print("Printing Record %s of Set %s | %s" % (counter1, counter2, city))
    counter1 +=1
    try:
        json_output = requests.get(query_url + city).json()
        lat.append(json_output["coord"]["lat"])
        long.append(json_output["coord"]["lon"])
        temp.append(json_output["main"]["temp_max"])
        humid.append(json_output["main"]["humidity"])
        cloudiness.append(json_output["clouds"]["all"])
        winds.append(json_output["wind"]["speed"])
        date.append(json_output["wind"]["speed"])
        country.append(json_output["sys"]["country"]) 
        #max_temp.append(json_outpu["main"]["temp_max"])
        city_name.append(city)        
        #print(counter1, counter2,city)
               
    except:
        print("City not found")      

        
    
    
```

```python
weather_df = pd.DataFrame({"City":city_name,
                          "Latitude": lat,
                          "Longitude": long,
                          "Temperature": temp,
                          "Humidity": humid,
                          "Cloudiness":cloudiness,
                          "Wind Speed": winds,
                          "Country": country,
                           "Date": date})
weather_df.head()
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
#temp vs lattitude linear regressions all below - pardon the formatting as they took up quite a bit of screen space
#keep scrolling down to see all...

x_values = weather_df['Latitude']
y_values = weather_df['Temperature']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,1)) + "x + " + str(round(intercept,1))

plt.scatter(weather_df['Latitude'], weather_df['Temperature'])
plt.title("City Temperature vs Latitude Regression")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6, 0),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel("Temperature")

print(f"rvalue    = {round(rvalue, 3)}")
print(f"r-squared = {round(rvalue**2, 3)}")
plt.savefig("Temp vs L Regress")
plt.show()
#keep scrolling - not sure why chart took up so much space
#Results show the highest temperatures on the center (near equator)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Max Temp  vs latitude in No hem. linear regression
x_values = Nor_hem_df['Latitude']
y_values = Nor_hem_df['Max Temp']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,1)) + "x + " + str(round(intercept,1))

plt.scatter(Nor_hem_df['Latitude'], Nor_hem_df['Max Temp'])
plt.title("Max Temp per City vs Latitude in Northern Hemisphere Regression")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6, 0),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel("Max Temp.")

print(f"rvalue    = {round(rvalue, 3)}")
print(f"r-squared = {round(rvalue**2, 3)}")
plt.savefig("Temp vs L NoH")

plt.show()
#Max Temp vs latitude in the Northern Hemisphere shows us that temperatures are warmer closer to the equater (lat of 0).
#see high negative R-value for this confirmation (temp's decrease as latitude increases). The r-squared is pretty high here
#as well so that confirms that the model is accurate in explaining the variability of the data.
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Max Temp  vs latitude in No hem. linear regression
x_values = Sou_hem_df['Latitude']
y_values = Sou_hem_df['Max Temp']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,1)) + "x + " + str(round(intercept,1))

plt.scatter(Sou_hem_df['Latitude'], Sou_hem_df['Max Temp'])
plt.title("Max Temp per City vs Latitude in Southern Hemisphere Regression")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6, 0),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel("Max Temp.")

print(f"rvalue    = {round(rvalue, 3)}")
print(f"r-squared = {round(rvalue**2, 3)}")
plt.savefig("Temp vs L SouH")
plt.show()
#Max temp / city in Southern hem. has a positive correlation where the higher the lattitude (less negative), the higher the temp.
#Worth noting is that there is less correlation in temperatures and latitude than there was in the northern hemisphere.
#the r squared being lower (relative to the r-squared in No hem) tells us there is more variability in the data not explained
#by the regression. Perhaps there is a higher frequency of micro-climates in the So Hemisphere as compared to the Northern Hemisphere
#due to topography or some other factor that potentially influences temperature.
```
